```python
import pandas as pd

prices = [
	(1, 2.12),
	(2, 2.56),
	(3, 3.10),
	(4, 3.16),
	(5, 3.58),
	(6, 5.12),
	(7, 5.16),
	(8, 5.20),
	(9, 4.12),
	(10, 4.10),
	(11, 3.65),
	(12, 4.25),

]

df = pd.DataFrame(prices, columns=["month", "prices"])
df = df.set_index("month")
df
```

```python
prices_usd = df['prices'].apply(lambda prices: prices/4)
df['prices'] = prices_usd
df
```
